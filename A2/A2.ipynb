## Task 1: Order of convergence [6 marks]

Halley's method is a root finding method, related to Newton's method, which uses the first and second derivatives of the function. To find a root of a function $f(z)$, starting with an initial guess $z_0$, Halley's method updates the guess until convergence is achieved. The Halley update is given by

$$
z_{k+1} = z_k - \frac{2 f(z_k) f'(z_k)}{2\left(f'(z_k)\right)^2 - f(z_k) f''(z_k)}.
$$

Perform a numerical investigation to determine the rate of convergence of Halley's method.

A general $N$-point finite difference approximation of the derivative $F' \left( x \right)$ of a sufficiently smooth function $F \left( x \right)$ can be written as

$$
F' \left( x \right) \approx \frac{1}{\Delta x} \sum_{i = 1}^N \alpha_i F \left( x + \beta_i \Delta x \right),
\qquad \qquad \qquad (*)
$$

$$

\alpha_1 = \frac{1}{2}, &\alpha_2 = -\frac{1}{2}, \\
\beta_1 = 1, &\beta_2 = -1,
\end{cases}
\qquad
\text{giving} \quad
F'(x) \approx \frac{1}{2\Delta x} \left(F\left(x + \Delta x\right) - F\left(x - \Delta x\right)\right).
$$

It is possible to find an $N$-point finite difference approximation which is at least $(N-1)$th order accurate, given any set of $N$ distinct $\beta_i$.

## Task 3: Bouncy-value problems [28 marks]

You've probably heard of initial-value problems and maybe boundary-value problems. But we're doing some pretty state-of-the-art stuff when considering *bouncy-value* problems. 

The idea is the following: You have a bouncy ball (e.g. a steel marble). You stand a given distance from a target (e.g. a tea cup) at a given height. You throw the ball from a given height. You have to pick a *horizontal speed* to throw the marble so that it lands in the cup.
You can bounce the ball on the floor either once or zero times but no more.

* The **goal** is to **find initial conditions** that will produce trajectories like the ones shown below.    

Below are two example solutions with zero and one bounce. In each case, the marble starts at a height of 150 cm. The teacup is 100 cm away and 75 cm above the ground. The blue lines show the trajectory. The dashed grey lines show the hypothetical continuation of the trajectory past the goal. 

![goal](goal.png)

We could model the following situation with a lot of various complications. However, we will simplify the problem as much as possible while still retaining some degree of realism.

### Mathematical model

All length units for this problem are in centimetres (cm), and time units are in seconds (s).

At a given horizontal position $x$, the **vertical** position of the ball $y(x)$ can be modelled by a second-order ordinary differential equation (ODE):

$$\large
u_{0}^{2}\, y''(x) \ = \ -g \left( \, 1 - e^{-(y(x) - a)/b} \, \right) - \kappa\, u_{0} \, y'(x), \qquad (1)
$$

with the following parameters:

- $g \approx 980$ cm$\cdot$s$^{-2}$ is the constant gravitational acceleration,
- $\kappa = 0.5$ s$^{-1}$ is a damping coefficient,
- $a = 1$ cm and $b = 0.25$ cm characterise the "bounciness" of the floor,
- $u_0$ is the chosen horizontal speed of the ball, in cm$\cdot$s$^{-1}$.

#### Initial conditions

This is a second-order ODE, so we need 2 initial conditions to fully specify the problem. A given ball trajectory is determined by the following initial conditions:

- the initial height of the ball $y(x=0) = y_0$, in cm,
- the initial vertical velocity $y'(x=0)$, in cm$\cdot$s$^{-1}$. For simplicity, we assume that we always throw in the horizontal direction, i.e. we always choose $y'(x=0) = 0$.

At any given point later, the instantaneous height of the ball, $y(x)$, is an implicit function of both $y_{0} > 0$ and $u_{0} > 0$. 

### Numerical solution

As we've done for similar problems in the course (see e.g. the Week 7 workshop task), we can build simulations of the ball trajectory by discretising the ODE (replacing the derivatives with finite difference approximations), choosing initial conditions, and marching the solution forward one time step at a time.

For simplicity, we assume that the horizontal speed of the ball is constant in time (and equal to $u_0$ cm$\cdot$s$^{-1}$). Therefore, a step size $\Delta x$ (in cm) has the same role as a time step $\Delta t$ -- in fact we have $\Delta x = u_0 \Delta t$.

Choose a step size $\Delta x$ (in cm). Let $y_n$ be an approximation to the solution of the ODE, at a given step $x_n = n\Delta x$, with $n = 0, 1, 2, \dots$, that is $y_n \approx y(x_n)$.

Here, we will use centred difference approximations for both $y'(x)$ and $y''(x)$:

$$
y'(x_n) \ \approx \ \frac{y(x + \Delta x) - y(x - \Delta x) }{2\Delta x} 
\approx \frac{y_{n+1} - y_{n-1} }{2\Delta x} \qquad (2a)
$$


$$
y''(x) \ \approx \ \frac{y(x + \Delta x) - 2y(x) + y(x - \Delta x) }{\Delta x^{2}} 
\approx \frac{y_{n+1} - 2y_n + y_{n-1} }{\Delta x^{2}} \qquad (2b)
$$

These formulas are straightforward provided $x \ne 0$. However, for $x=0$, we have a problem. Thankfully, we can use our initial conditions to solve it:

$$
y'(0) \ = \ 0 \qquad \implies \qquad y_{-1} \ = \ y_1.
$$

We can substitute the above into the second order finite difference approximation at $x=0$, to obtain

$$
y''(0) \ \approx \ 2\,\frac{y_1 - y_0}{\Delta \, x^{2}}. \qquad (3)
$$

Using all of the above information, you can write an iteration scheme, to calculate $y_{n+1}$ as a function of previously computed values $y_n$ and $y_{n-1}$, for $n=2, 3, 4, \dots$.

- At $n=0$, the computed solution is simply $y_0$, your chosen initial height.
- At $n=1$, you can obtain an expression for $y_1$ as a function of $y_0$ and $y''(0)$, by combining $(1)$ and $(3)$, with your chosen horizontal velocity $u_0$, and recalling that $y'(0) = 0$.
- At $n \geq 2$, you can obtain an expression for $y_{n+1}$ as a function of $y_n$ and $y_{n-1}$ by substituting $(2a)$ and $(2b)$ into $(1)$ and rearranging the expression.

### Throwing the marble into the cup

Now we want to throw the marble in the teacup. We can write this problem as a root-finding exercise.

If the teacup is located at $(x_{\text{teacup}}, y_{\text{teacup}})$, for a given initial throwing height $y_0$, we want to find the horizontal speed $u_0$ such that the trajectory of the ball lands it into the cup.

Implicitly, a given trajectory depends on many more parameters than we usually denote. In this case,

$$
y(x) = y(x\,; \, u_{0}, y_{0}, g, \kappa, a, b).
$$

We typlically omit all parametres other than $x$, however the trajectory is just as much a valid function of any other parameters. 

We can therefore define a residual function $F$ that measures how close a trajectory is to a given point as a function of any one of the implicity parameters. 

Fixing, $y_{0},g,\kappa,a,b$ and $x_{\text{teacup}}$, $y_{\text{teacup}}$,

$$
F(u_{0})  \ = \   y(x_{\text{teacup}}\,; \, u_{0}, y_{0}, g, \kappa, a, b) - y_{\text{teacup}}.
$$

We can then try to find values for $u_{0}$ where 

$$
F(u_{0}) = 0.
$$

Picking an initial parameter so that a trajectory lands at a given location may not look like a root-finding exercise, but it is if you make the implicit parameters clear.

#### Strategy

In the bounce case, the interaction with the floor produces unpredictable results. However, we know that friction always reduces the motion, and the interaction with the floor produces most of the damping (also true with real objects). 

However, we can pretend there is exactly zero damping, which gives us an estimate that will *undershoot* the mark when we include friction. 

In this case, we estimate that the marble falls to the floor with a parabolic trajectory. It then bounces back and returns to its original height with a mirror-image trajectory. After that, the marble continues to fall toward the target but with a reduced distance. 

Putting everything together, produces a lower bound for the speed estimate. The zero-bounce estimate from Task 3c now gives us an upper bound, and we can bracket our solution with:

$$
\frac{x_{\text{teacup}} \sqrt{\frac{g}{2\,h}}}{1 + 2\sqrt{\frac{y_0}{h}}} < u_0 < x_{\text{teacup}}\sqrt{\frac{g}{2\,h}}, \qquad \text{where} \qquad h = y_0 - y_{\text{teacup}}.
$$

Starting from this interval $[u_{\min}, u_{\max}]$, it's possible to use a modified version of the bisection method, which also **counts the total number of bounces and peaks** for the trajectory with the midpoint velocity $u_{\text{midpoint}} = \frac{u_{\min} + u_{\max}}{2}$.

You can then choose the next half-interval based on this information. The constraints are as follows:

- The marble must bounce **exactly once** before getting into the cup.
- The marble must **fall downwards** into the cup.

### Appendix: the physics of the bouncy ball

_This is not part of the assignment. If you're interested in how the ODE (1) is derived, read along!_

The ball moves in the two-dimensional plane $(\,x(t),\, y(t)\,)$, where $x$ is *horizontal*, $y$ is *vertical* and $t$ is *time*. For simplicity, we assume the horizontal speed of the ball is constant in time. Therefore, the horizontal position and time follow a simple linear relationship,

$$
x \ = \ u_{0}\, t,
$$

where $u_0$ is the chosen horizontal speed of the ball (cm/s). We can, therefore, eliminate the explicit dependence on time when considering the ball's vertical motion, by using the chain rule to relate all time derivatives to the derivative with respect to the horizontal position,

$$
\frac{d}{dt} \ = \ u_{0}\, \frac{d}{dx}.
$$

The evolution over time of the vertical position of the ball can then be modelled with a system of 2 differential equations:

$$
\begin{align}
&u_{0}\, y'(x) = v(x) & (1)\\
&u_{0}\, v'(x) \ = \ -g \left( \, 1 - e^{-(y(x) - a)/b} \, \right) - \kappa\, v(x) & (2)
\end{align}
$$


* (1) is the definition of vertical velocity as a function of horizontal position.
* (2) is Newton's Law of Motion for the forces acting on the ball.
    - The first term in (2), $-g$, is the gravitational force exerted on the ball.
    - The second term, $g e^{-(y(x) - a)/b}$, models the force exerted by the floor onto the ball when it bounces.
    - The third term, $-\kappa v(x)$, represents the damping exerted on the ball which decelerates it.

The exact value and form of the damping don't matter as much as the fact that it's simply there.

The bouncing force is a simple model for the ball bouncing off the floor, hence the exponential force term. This term is minuscule unless the ball is very close to the floor, in which case it becomes significant extremely quickly as the ball progresses further downward. The physics is quite rough; however, a more sophisticated model would act qualitatively the same way.
